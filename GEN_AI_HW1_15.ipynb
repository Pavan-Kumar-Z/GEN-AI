```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**4 - 3*x**3 + 2

def grad_f(x):
    return 4*x**3 - 9*x**2

def gradient_descent(starting_x, learning_rate, num_iterations, max_value=1e6):
    x = starting_x
    history = [x]
    for _ in range(num_iterations):
        gradient = grad_f(x)
        x = x - learning_rate * gradient
        # Safeguard: Stop if x becomes too large
        if abs(x) > max_value:
            print(f"Divergence detected for η = {learning_rate}. Stopping early.")
            break
        history.append(x)
    return history

```

```python
starting_x = 1.0
num_iterations = 50
learning_rates = [0.1, 1.0, 10.0]

results = {}
for lr in learning_rates:
    results[lr] = gradient_descent(starting_x, lr, num_iterations)

plt.figure(figsize=(10, 6))
for lr, history in results.items():
    plt.plot(history, label=f"η = {lr}")
plt.xlabel("Iteration")
plt.ylabel("x")
plt.title("Gradient Descent Convergence for Different Learning Rates")
plt.legend()
plt.grid()
plt.show()
```
