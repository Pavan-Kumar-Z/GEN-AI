```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)
```

```python
x = np.linspace(-5, 5, 400)

sigmoid_vals, sigmoid_deriv = sigmoid(x), sigmoid_derivative(x)
tanh_vals, tanh_deriv = tanh(x), tanh_derivative(x)
relu_vals, relu_deriv = relu(x), relu_derivative(x)
leaky_relu_vals, leaky_relu_deriv = leaky_relu(x), leaky_relu_derivative(x)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Activation Functions and Their Derivatives", fontsize=14, fontweight="bold")

# Define vanishing gradient regions (shaded in red)
vanish_region = (-0.1, 0.1)

# Sigmoid
axes[0, 0].plot(x, sigmoid_vals, label="Sigmoid", color='b')
axes[0, 0].fill_between(x, sigmoid_deriv, where=(sigmoid_deriv < vanish_region[1]), color='r', alpha=0.3)
axes[0, 0].set_title("Sigmoid")
axes[0, 0].legend()

axes[1, 0].plot(x, sigmoid_deriv, label="Sigmoid Derivative", color='g')
axes[1, 0].fill_between(x, sigmoid_deriv, where=(sigmoid_deriv < vanish_region[1]), color='r', alpha=0.3)
axes[1, 0].set_title("Sigmoid Derivative (Vanishing Gradient Highlighted)")
axes[1, 0].legend()

# Tanh
axes[0, 1].plot(x, tanh_vals, label="Tanh", color='b')
axes[0, 1].fill_between(x, tanh_deriv, where=(tanh_deriv < vanish_region[1]), color='r', alpha=0.3)
axes[0, 1].set_title("Tanh")
axes[0, 1].legend()

axes[1, 1].plot(x, tanh_deriv, label="Tanh Derivative", color='g')
axes[1, 1].fill_between(x, tanh_deriv, where=(tanh_deriv < vanish_region[1]), color='r', alpha=0.3)
axes[1, 1].set_title("Tanh Derivative (Vanishing Gradient Highlighted)")
axes[1, 1].legend()

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("ReLU and Leaky ReLU Activation Functions", fontsize=14, fontweight="bold")

# ReLU
axes[0, 0].plot(x, relu_vals, label="ReLU", color='b')
axes[0, 0].set_title("ReLU")
axes[0, 0].legend()

axes[1, 0].plot(x, relu_deriv, label="ReLU Derivative", color='g')
axes[1, 0].fill_between(x, relu_deriv, where=(x < 0), color='r', alpha=0.3)
axes[1, 0].set_title("ReLU Derivative (Zero Gradient for x<0)")
axes[1, 0].legend()

# Leaky ReLU
axes[0, 1].plot(x, leaky_relu_vals, label="Leaky ReLU", color='b')
axes[0, 1].set_title("Leaky ReLU")
axes[0, 1].legend()

axes[1, 1].plot(x, leaky_relu_deriv, label="Leaky ReLU Derivative", color='g')
axes[1, 1].set_title("Leaky ReLU Derivative (No Vanishing Gradient)")
axes[1, 1].legend()

plt.show()
```
